```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def poly_filler(array, x_array, y_array, x_points, y_points, val=1, clockwise=True):
    x2_points = np.append(x_points[1:], x_points[0])
    y2_points = np.append(y_points[1:], y_points[0])
    x_grid = np.array([x_array for i in y_array])
    y_grid = np.array([y_array for i in x_array]).transpose()

    where = np.ones_like(array)
    for x1, x2, y1, y2 in zip(x_points, x2_points, y_points, y2_points):
        if x2 == x1:
            continue
        m = (y2 - y1) / (x2 - x1)
        c = y1 - (x1 * m)
        y_func_grid = (m * x_grid + c) - y_grid
        if x2 > x1:
            where *= y_func_grid >= 0
        else:
            where *= y_func_grid <= 0
    array[where == 1] = val
    return array
```
